Choose any data attribute with 1000 instance. Test normality of data by using Kurtosis. Apply transformation to convert data to normal. ( e^x , 1/x , log(x)).

```python
#importing libs..

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline

# Set style
plt.style.use('ggplot')
```

```python
import numpy as np                                                                       
from scipy.stats import kurtosis                                                         

min_range = 1.0                                                                         
max_range = 10.0                                                                         
midpoint = (max_range + min_range)/2                                                     
samples = 1000                                                                      

#generate 1000 instance uniform attribute.
def filter_tails(x):                                                                     
    return x[(x >= min_range) & (x <= max_range)]                                        

run_1000 = np.random.uniform(min_range, max_range, samples)                                 
```

```python
len(run_1000) #1000 instances
```

```python
print(run_1000)
```

```python
plt.hist(run_1000)
plt.title('Histogram for: 1000 instances')
plt.suptitle('Randomly Sampled 1000 Uniform Variables', fontsize=14)
plt.show()
```

```python
# apply kurtosis test

import scipy.stats as stats
from scipy.stats import kurtosistest
kur_run = kurtosistest(run_1000)
```

```python
kur_run
```

Here Kurtosis value is too high as data is not normal.

```python
stat_run=kur_run[0]
pvalue_run=kur_run[1]
```

```python
import scipy.stats as stats
from scipy.stats import kurtosis

def accept_reject(kur_run):
    if pvalue_run > .05 :
        print ("Don't reject normality hypothesis")
    else:
        print ("Reject normality hypothesis")
        
accept_reject(stats.kurtosis(run_1000))
```

Normal Data Kurtosis needed to be nearby zero. This have pvalue > 0.05. so it is rejected

# Transformation

# transformation of data using e^x

```python
x_e= np.exp(run_1000)
x_e
```

```python
kur_e = kurtosistest(x_e)
print(kur_e)
stat_e=kur_e[0]
pvalue_e = kur_e[1]
```

```python
plt.hist(x_e)
plt.title('Histogram for:e^x 1000 instances')
plt.suptitle('e^x 1000 Variables', fontsize=14)
plt.show()
```

# Transformation of data using 1/x

```python

x = 1/run_1000
x
```

```python
kur_x = kurtosistest(x)
print(kur_x)
stat_x=kur_x[0]
pvalue_x = kur_x[1]
```

```python
plt.hist(x)
plt.title('Histogram for:1/x 1000 instances')
plt.suptitle('1/x 1000 Variables', fontsize=14)
plt.show()
```

# Transformation of data using log(x)

```python

x_log=np.log(run_1000)
x_log
```

```python
kur_log = kurtosistest(x_log)
print(kur_log)
stat_log=kur_log[0]
pvalue_log = kur_log[1]
```

```python
pvalue_log
```

Here we can see that pvalue is near by zero . 

```python
plt.hist(x_log)
plt.title('Histogram for:log(x) 1000 instances')
plt.suptitle('log(x) 1000 Variables', fontsize=14)
plt.show()
```

# We can see here that trasformation using logarithmic function gives a kurtosis nearer to zero. This trasformation is most suitable for uniform kind of instances.
